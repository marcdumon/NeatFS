# Directory Filter

```python
import pandas as pd
```

## Load the fs index

```python
df = pd.read_csv('../data/fs_index.csv', low_memory=False)
df.head()
```

## All parent directories

There are more than 1.3 million directories!

```python
parents = sorted(df['parent'].unique())
print('Nr. of parents:', len(parents))
parents[:5], parents[-5:]
```

## Filter directory 

```python
dir = '/mnt/Data/obsidian'
```

### All directories

```python
mask = df['parent'].str[:len(dir)] == dir
df[mask]
```

```python
sorted(df[mask]['parent'].unique())
```

### Directories 1 level deeper

```python
parent_level = len(dir.split('/')) -1
parent_level
```

```python
mask = (df['parent'].str[:len(dir)] == dir) & (df['parent'].str.count('/') == parent_level +1)
df[mask]
```
